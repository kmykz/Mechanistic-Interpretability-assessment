```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModel
from transformers.models.bert.configuration_bert import BertConfig
from tqdm.auto import tqdm # Import tqdm

# ======================
# Setup
# ======================
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {DEVICE}")

# DNABERT tokenizer & backbone
MODEL_NAME = "zhihan1996/DNABERT-2-117M"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, trust_remote_code=True)

config = BertConfig.from_pretrained(MODEL_NAME)
backbone = AutoModel.from_pretrained(MODEL_NAME, trust_remote_code=True, config=config)


# ======================
# Dataset class (same style as before)
# ======================
class ClinVarDataset(Dataset):
    def __init__(self, records, tokenizer, use_alt=True, max_len=512):
        self.records = records
        self.tokenizer = tokenizer
        self.use_alt = use_alt
        self.max_len = max_len
        self.label2id = {
            "Benign": 0,
            "Likely_benign": 1,
            "Pathogenic": 2,
            "Likely_pathogenic": 3
        }

    def __len__(self):
        return len(self.records)

    def __getitem__(self, idx):
        ref, alt, var_type = self.records[idx][0]
        label = self.records[idx][1]

        seq = alt if self.use_alt else ref  # choose which sequence to use

        enc = self.tokenizer(
            seq,
            padding="max_length",
            truncation=True,
            max_length=self.max_len,
            return_tensors="pt"
        )
        return {
            "input_ids": enc["input_ids"].squeeze(0),
            "attention_mask": enc["attention_mask"].squeeze(0),
            "label": torch.tensor(self.label2id[label]),
        }

# ======================
# Classifier Model
# ======================
class DNABERTClassifier(nn.Module):
    def __init__(self, backbone, num_classes,dropout_rate=0.3):
        super().__init__()
        self.backbone = backbone
        hidden_size = backbone.config.hidden_size
        self.classifier = nn.Linear(hidden_size, num_classes)

        # Freeze backbone parameters
        for param in self.backbone.parameters():
            param.requires_grad = False

    def forward(self, input_ids, attention_mask=None):
        outputs = self.backbone(input_ids=input_ids, attention_mask=attention_mask)
        hidden_states = outputs[0]  # Access last hidden state from the tuple
        #pooled = torch.mean(hidden_states, dim=1)  # mean pooling
        pooled = hidden_states[:, 0, :]
        logits = self.classifier(pooled)
        return logits


train_dataset = ClinVarDataset(data, tokenizer)
train_loader = DataLoader(train_dataset, batch_size=256, shuffle=True)

model = DNABERTClassifier(backbone, num_classes=4).to(DEVICE)
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.classifier.parameters(), lr=2e-4)

for epoch in range(3):
    model.train()
    total_loss = 0
    predictions = []
    true_labels = []
    for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}"):
        input_ids = batch["input_ids"].to(DEVICE)
        attention_mask = batch["attention_mask"].to(DEVICE)
        labels = batch["label"].to(DEVICE)

        optimizer.zero_grad()
        logits = model(input_ids, attention_mask)
        loss = criterion(logits, labels)
        loss.backward()
        optimizer.step()
        preds = torch.argmax(logits, dim=1)

        total_loss += loss.item()
        predictions.extend(preds.cpu().tolist())
        true_labels.extend(labels.cpu().tolist())

    avg_loss = total_loss / len(train_loader)
    print(f"Epoch {epoch+1}: Loss = {avg_loss:.4f}")
    #accuracy = accuracy_score(true_labels, predictions)
    #print(f"Accuracy: {accuracy:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
indx_pic = 0
def visualize_attention(model, tokenizer, sequence, layer=0, head=0):
    global indx_pic
    model.backbone.config.output_attentions = True
    enc = tokenizer(sequence, return_tensors="pt", padding="max_length", truncation=True, max_length=64)
    input_ids = enc["input_ids"].to(DEVICE)
    attention_mask = enc["attention_mask"].to(DEVICE)

    with torch.no_grad():
        outputs = model.backbone(
            input_ids=input_ids,
            attention_mask=attention_mask,
            output_attentions=True,
            output_hidden_states=True
        )

    attentions = outputs[2] if len(outputs) > 2 else None

    return


    attn = attentions[layer][0, head].cpu().tolist()

    tokens = tokenizer.convert_ids_to_tokens(input_ids[0])
    seq_len = attentions[layer].size(-1)
    tokens = tokens[:seq_len]


    plt.figure(figsize=(10, 8))
    sns.heatmap(attn, xticklabels=tokens, yticklabels=tokens, cmap="viridis")
    plt.title(f"Layer {layer}, Head {head} Attention")
    plt.xlabel("Key Tokens")
    plt.ylabel("Query Tokens")
    #plt.show()
    plt.savefig(f"Images/attention_layer_{layer}_head_{head}_{indx_pic}.png")
    indx_pic += 1
    plt.close()
```

```python
def get_hidden_and_logits(model, tokenizer, seq, max_len=128):
    inputs = tokenizer(seq, return_tensors="pt", padding="max_length",
                       truncation=True, max_length=max_len).to(next(model.parameters()).device)
    with torch.no_grad():
        outputs = model.backbone(
            input_ids=inputs["input_ids"],
            attention_mask=inputs["attention_mask"],
            output_hidden_states=True
        )
        hidden = outputs[0]
        pooled = hidden[:, 0, :]
        logits = model.classifier(pooled)
    return hidden, logits, inputs
def patch_variant_region(model, clean_seq, corrupt_seq, tokenizer, variant_start, variant_end, window=5, max_len=128):


    clean_hidden, clean_logits, clean_inputs = get_hidden_and_logits(model, tokenizer, clean_seq, max_len)
    corrupt_hidden, corrupt_logits, corrupt_inputs = get_hidden_and_logits(model, tokenizer, corrupt_seq, max_len)

    clean_pred = torch.argmax(clean_logits, dim=-1).item()
    corrupt_pred = torch.argmax(corrupt_logits, dim=-1).item()
    if (clean_pred == corrupt_pred):
        return {
            "clean_pred": clean_pred,
            "corrupt_pred": corrupt_pred,
            "patched_pred": clean_pred,
            "patched_range": None
        }
    print(f"Clean prediction: {clean_pred}, Corrupt prediction: {corrupt_pred}")


    variant_token_start = max(1, variant_start - window)
    variant_token_end = min(clean_hidden.size(1)-1, variant_end + window)

    print(f"Patching tokens {variant_token_start}:{variant_token_end}")

    patched_hidden = clean_hidden.clone()
    patched_hidden[:, variant_token_start:variant_token_end, :] = \
        corrupt_hidden[:, variant_token_start:variant_token_end, :]

    pooled = patched_hidden[:, 0, :]
    patched_logits = model.classifier(pooled)
    patched_pred = torch.argmax(patched_logits, dim=-1).item()
    if (patched_pred == corrupt_pred):
        print("Patched prediction matches corrupt prediction!")
    print(f"Patched prediction (variant region): {patched_pred}")

    return {
        "clean_pred": clean_pred,
        "corrupt_pred": corrupt_pred,
        "patched_pred": patched_pred,
        "patched_range": (variant_token_start, variant_token_end)
    }
```

```python
import re

def corrupt_creb_site(sequence, method='mutate_core'):

    creb_motif = 'TGACGTCA'
    match = re.search(creb_motif, sequence, re.IGNORECASE)

    if not match:
        #print(f"Warning: CREB motif '{creb_motif}' not found in the sequence.")
        return sequence,-1

    start, end = match.span()
    original_site = match.group(0)

    corrupted_sequence = list(sequence)

    if method == 'mutate_core':
        corrupted_sequence[start + 3] = 'T'
        corrupted_sequence[start + 4] = 'T'
        corruption_result = "TGATTCA"

    elif method == 'break_palindrome':
        corrupted_sequence[start + 5] = 'A'
        corrupted_sequence[start + 6] = 'A'
        corruption_result = "TGACGACA"

    elif method == 'replace_neutral':
        neutral_site = 'ATATATAT'
        for i in range(len(neutral_site)):
            corrupted_sequence[start + i] = neutral_site[i]
        corruption_result = neutral_site

    else:
        raise ValueError(f"Unknown method '{method}'. Choose from 'mutate_core', 'break_palindrome', or 'replace_neutral'.")

    print(f"Method '{method}': Changed '{original_site}' at index {start} to '{corruption_result}'")
    st = "".join(corrupted_sequence)
    return st,start

random_indices = random.sample(range(len(data)), 10000)
for ind in random_indices:
    clean_seq = data[ind][0][1]
    corrupt_seq,variant_start = corrupt_creb_site(clean_seq,method='replace_neutral')
    if variant_start ==-1:
        continue
    variant_end = variant_start + 6
    if variant_start != -1:
        continue
    res = patch_variant_region(model, clean_seq, corrupt_seq, tokenizer,
                           variant_start, variant_end, window=5, max_len=128)
```
